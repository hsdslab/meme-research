```python
import pandas as pd
from tqdm.notebook import tqdm
import os

print(os.getcwd())

with open('data/phashes/imgflip_all_phashes.txt','r') as f:
    imgflip = f.readlines()

path_phash_map = {}

for line in tqdm(imgflip):
    path, phash = line.strip().split('\t')
    path_phash_map[path] = phash

path_phash_df = pd.DataFrame(list(path_phash_map.items()), columns=['path','phash'])

meme_df = pd.read_parquet('data/meme_entries.parquet')
meme_df
```

```python
path_phash_df['path'] = path_phash_df['path'].apply(lambda x: x.replace('\\','/'))
meme_df['path'] = meme_df['path'].apply(lambda x: x.replace('\\','/'))
path_phash_df
```

```python
merged = meme_df.merge(path_phash_df, on='path', how='right')
merged
```

```python
merged.to_parquet('data/meme_entries.parquet')
```
