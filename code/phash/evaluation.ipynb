Create a dict to store the assigned templates for each cluster.

```python
import os 
import json
import pandas as pd
import pathlib
```

```python

clustering_selection = json.load(open('imgflip_clustering_selection.json'))
clustering_selection
cluster_template_dict = {}

keys = clustering_selection.keys()

for key, value in clustering_selection.items():
    cluster_no = key.split('-')[0]
    filepath = "-".join(key.split('-')[1:])
    path = pathlib.PurePath(filepath)
    template = path.parent.name.lower()
    if cluster_no in cluster_template_dict:
        cluster_template_dict[cluster_no].append((template,value))
    else:
        cluster_template_dict[cluster_no] = [(template,value)]
cluster_template_dict
```

For each cluster, we only keep the template that is the closest to the cluster medoid.

```python
cluster_final_pred = {}
for key, value in cluster_template_dict.items():
    cluster_final_pred[int(key)] = min(value, key=lambda x: x[1])[0]

cluster_final_pred
```

```python
clusters_file =  r"D:\Murgi\code\memes2024\meme-research-2024\clustering_output.txt"

with open(clusters_file,'r') as f:
    lines = f.readlines()

# create a pandas dataframe
list_of_dicts = [json.loads(line) for line in lines]
df = pd.DataFrame.from_dict(list_of_dicts)
df.columns = ['cluster_no','images','medoid_phash','medoid_path']
df
```

```python
cluster_pred_df = pd.DataFrame.from_dict(cluster_final_pred, orient='index', columns=['predicted_template'])
cluster_pred_df.index.name = 'cluster_no'
cluster_pred_df
```

```python
df = df.merge(cluster_pred_df, left_on='cluster_no', right_index=True)
df
```

```python
new_df = df.explode('images')
new_df = new_df[['images', 'predicted_template']]
new_df['predicted_template'] = new_df['predicted_template'].apply(lambda x: x.lower())

new_df.to_parquet(r"D:\Murgi\code\memes2024\meme-research-2024\data\phashes\final_preds_imgflip_2024.parquet", index=False)
```

```python
def merge_columns_by_prefix(df, prefix, new_column_name, drop_old=True):
    """
    Merges multiple columns in a DataFrame that start with a given prefix into a single column.
    
    Parameters:
        df (pd.DataFrame): The DataFrame to operate on.
        prefix (str): The prefix to identify the columns to merge.
        new_column_name (str): The name of the new column that will contain the merged data.
        drop_old (bool): Whether to drop the old columns after merging. Default is True.
    
    Returns:
        pd.DataFrame: The DataFrame with the merged column.
    """
    # Identify columns that start with the prefix
    columns_to_merge = [col for col in df.columns if col.startswith(prefix)]
    
    # Initialize the new column with the first relevant column found
    if columns_to_merge:
        df[new_column_name] = df[columns_to_merge[0]]
        
        # Sequentially fill in the new column with non-null values from other columns
        for col in columns_to_merge[1:]:
            df[new_column_name] = df[new_column_name].fillna(df[col])
        
        # Optionally drop the original columns
        if drop_old:
            df.drop(columns=columns_to_merge, inplace=True)
    else:
        print(f"No columns start with the prefix '{prefix}'. No changes made.")
    
    return df

merged = merge_columns_by_prefix(merged,"id_","id")
merged = merge_columns_by_prefix(merged,"source_","source")
merged
```

# Calculate metrics and store them on MLFlow

```python
memes_df = pd.read_parquet(r'D:\Murgi\code\memes2024\meme-research-2024\data\meme_entries.parquet')
memes_df 
```

```python
# encode the template to a number
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(memes_df["template_name"])
memes_df["template_id"] = le.transform(memes_df["template_name"])
memes_df.loc[:,'path'] = memes_df.loc[:,'path'].apply(lambda x: x.replace('\\','/'))
memes_df
```

```python
new_df["predicted_template_id"] = le.transform(new_df["predicted_template"])
new_df.loc[:,'images'] = new_df.loc[:,'path'].apply(lambda x: x.replace('\\','/'))

new_df
```

```python
eval_df = new_df.merge(memes_df, left_on="images", right_on="path", how="right")
eval_df
```

Without taking into account the outliers

```python
from sklearn.metrics import f1_score,matthews_corrcoef,cohen_kappa_score

wo_templateless = eval_df.dropna(subset=['predicted_template_id'])
y_true = wo_templateless["template_id"]
y_pred = wo_templateless["predicted_template_id"]


f1 = f1_score(y_true, y_pred, average='weighted')
mcc = matthews_corrcoef(y_true, y_pred)
kappa = cohen_kappa_score(y_true, y_pred)
# gmean = geometric_mean_score(y_true, y_pred, average='weighted')

print(f"F1 (weighted): {f1}")
print(f"MCC: {mcc}")
print(f"Kappa: {kappa}")
# print(f"G-Mean (weighted): {gmean}")
```

With the outliers

```python
from sklearn.metrics import f1_score,matthews_corrcoef,cohen_kappa_score

with_templateless = eval_df.fillna(-1)
y_true = with_templateless["template_id"]
y_pred = with_templateless["predicted_template_id"]


f1 = f1_score(y_true, y_pred, average='weighted')
mcc = matthews_corrcoef(y_true, y_pred)
kappa = cohen_kappa_score(y_true, y_pred)
# gmean = geometric_mean_score(y_true, y_pred, average='weighted')

print(f"F1 (weighted): {f1}")
print(f"MCC: {mcc}")
print(f"Kappa: {kappa}")
# print(f"G-Mean (weighted): {gmean}")
```

```python
import constants
import mlflow

mlflow.set_tracking_uri('https://dagshub.com/levente-murgas/meme-research-2024.mlflow')
# os.environ['MLFLOW_TRACKING_URI']=f"'https://dagshub.com/levente-murgas/meme-research-2024.mlflow'"

# # Recommended to define as environment variables
os.environ['MLFLOW_TRACKING_USERNAME'] = constants.MLFLOW_TRACKING_USERNAME
os.environ['MLFLOW_TRACKING_PASSWORD'] = constants.MLFLOW_TRACKING_PASSWORD
```

```python
def get_experiment_id(name):
    exp = mlflow.get_experiment_by_name(name)
    if exp is None:
      exp_id = mlflow.create_experiment(name)
      return exp_id
    return exp.experiment_id

exp_id = get_experiment_id("phash_clustering_imgflip_2024")
print(exp_id)
```

```python
from sklearn.cluster import HDBSCAN

CLUSTERING_MIN_SAMPLES=5
hdbscan = HDBSCAN(metric='hamming',store_centers='medoid', n_jobs=-1, min_samples=CLUSTERING_MIN_SAMPLES)

with  mlflow.start_run(experiment_id=exp_id):
    mlflow.log_params({
        "metric": 'hamming',
        "min_samples": CLUSTERING_MIN_SAMPLES,
    })


    mlflow.log_metrics({
        "f1weighted": f1,
        "mcc": mcc,
        "kappa": kappa
    })

    model_info = mlflow.sklearn.log_model(
        sk_model=hdbscan,
        artifact_path="hdbscan_phash",
        registered_model_name="hdbscan_phash",
    )
```
